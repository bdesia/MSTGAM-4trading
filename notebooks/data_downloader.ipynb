```python
from datetime import date
from utils import download_ticker

# List of stocks to download data for
tickers_list = [
    'NVDA', 'AVGO', 'AAPL', 'META', 'PLTR', 'AMD', 'MU', 'ORCL', 'INTC',
    'AMZN', 'MELI', 'BABA', 
    'TLSA', 'NIO', 
    'DIS', 
    'BITF', 'COIN', 'HUT', 'RIOT',
    'UBER', 'GLOB',
    'RGTI',
    'SATL',
        ]

today_str = date.today().strftime('%Y-%m-%d')
start_str ='2015-10-17'

data = {}
for ticker in tickers_list:
    print(f"Downloading data for {ticker}...")
    data[ticker] = download_ticker(ticker, start = start_str, end = today_str, filename = f'{ticker}.csv')
    print(f"Data for {ticker} downloaded successfully.")
    print("-" * 40)
```
